#**Thuật toán Bubble Sort**



```python
def bubble_sort(nums):  
    # đặt swapped thành True để vòng lặp trông chạy ít nhất một lần
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:
                # Hoán đổi các yếu tố
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                
                swapped = True

list1 = [1, 4, 7]
list2 = [1, 3 ,4]
list3 = [2 ,6]
list = list1 + list2 + list3
bubble_sort(list)  
print(list) 
```

#**Thuật toán Selection Sort**

```python
def selection_sort(nums):  
    # Giá trị này của i tương ứng với số lượng giá trị đã được sắp xếp
    for i in range(len(nums)):
        # Ta cho rằng mục đầu tiên của phân khúc chưa phân loại là nhỏ nhất
        lowest_value_index = i
        # Vòng lặp này lặp lại trên các mục chưa phân loại
        for j in range(i + 1, len(nums)):
            if nums[j] < nums[lowest_value_index]:
                lowest_value_index = j
        # Giá trị hoán đổi của phần tử chưa phân loại thấp nhất với phần tử chưa phân loại đầu tiên
        # yếu tố
        nums[i], nums[lowest_value_index] = nums[lowest_value_index], nums[i]


list1 = [1, 4, 7]
list2 = [1, 3 ,4]
list3 = [2 ,6]
list = list1 + list2 + list3 
selection_sort(list)  
print(list)  
```

#**Thuật toán Insertion Sort**

```python
def insertion_sort(nums):  
    # Bắt đầu trên phần tử thứ hai khi ta cho rằng phần tử đầu tiên được sắp xếp
    for i in range(1, len(nums)):
        item_to_insert = nums[i]
        # Và giữ một tham chiếu của chỉ mục của phần tử trước đó
        j = i - 1
        # Di chuyển tất cả các mục của phân đoạn đã sắp xếp về phía trước nếu chúng lớn hơn
        # mục cần chèn
        while j >= 0 and nums[j] > item_to_insert:
            nums[j + 1] = nums[j]
            j -= 1
        # Chèn mục
        nums[j + 1] = item_to_insert
        
list1 = [1, 4, 7]
list2 = [1, 3 ,4]
list3 = [2 ,6]
list = list1 + list2 + list3 
insertion_sort(list)  
print(list)  
```

#**Thuật toán Heap Sort**

```python
def heapify(nums, heap_size, root_index):  
    # Giả sử chỉ mục của phần tử lớn nhất là chỉ mục gốc
    largest = root_index
    left_child = (2 * root_index) + 1
    right_child = (2 * root_index) + 2

    # Nếu con bên trái của gốc là một chỉ mục hợp lệ và phần tử lớn hơn
    # so với phần tử lớn nhất hiện tại, sau đó cập nhật phần tử lớn nhất
    if left_child < heap_size and nums[left_child] > nums[largest]:
        largest = left_child

    # Làm tương tự 
    if right_child < heap_size and nums[right_child] > nums[largest]:
        largest = right_child

    # Nếu phần tử lớn nhất không còn là phần tử gốc, hãy trao đổi chúng
    if largest != root_index:
        nums[root_index], nums[largest] = nums[largest], nums[root_index]
        # Chất đống các yếu tố gốc mới để đảm bảo nó là lớn nhất
        heapify(nums, heap_size, largest)


def heap_sort(nums):  
    n = len(nums)

    # Tạo Một heap  Tối đa từ danh sách
    # Đối số thứ hai của phạm vi có nghĩa là chúng ta dừng lại ở phần tử trước -1 tức là.
    # Yếu tố đầu tiên của danh sách.
    # Đối số thứ 3 của phạm vi có nghĩa là chúng ta lặp lại ngược lại, giảm số lượng
    # của i by 1
    for i in range(n, -1, -1):
        heapify(nums, n, i)

    # Di chuyển gốc của heap  tối đa đến cuối
    for i in range(n - 1, 0, -1):
        nums[i], nums[0] = nums[0], nums[i]
        heapify(nums, i, 0)


list1 = [1, 4, 7]
list2 = [1, 3 ,4]
list3 = [2 ,6]
list = list1 + list2 + list3 
heap_sort(list)  
print(list)  
```

#**Thuật toán Merge Sort**

```python
def merge(left_list, right_list):  
    sorted_list = []
    left_list_index = right_list_index = 0

    # Ta sử dụng độ dài danh sách thường xuyên, vì vậy nó tiện dụng để tạo biến
    left_list_length, right_list_length = len(left_list), len(right_list)

    for _ in range(left_list_length + right_list_length):
        if left_list_index < left_list_length and right_list_index < right_list_length:
            # Ta kiểm tra giá trị nào từ đầu mỗi danh sách nhỏ hơn
            # Nếu mục ở đầu danh sách bên trái nhỏ hơn, hãy thêm mục đó
            # vào danh sách đã sắp xếp
            if left_list[left_list_index] <= right_list[right_list_index]:
                sorted_list.append(left_list[left_list_index])
                left_list_index += 1
            # Nếu mục ở đầu danh sách bên phải nhỏ hơn, hãy thêm mục đó
            # vào danh sách đã sắp xếp
            else:
                sorted_list.append(right_list[right_list_index])
                right_list_index += 1

        # Nếu ta đã đến cuối danh sách bên trái, hãy thêm các phần tử
        # từ danh sách bên phải
        elif left_list_index == left_list_length:
            sorted_list.append(right_list[right_list_index])
            right_list_index += 1
        # Nếu ta đã kết thúc danh sách bên phải, hãy thêm các yếu tố
        # từ danh sách bên trái
        elif right_list_index == right_list_length:
            sorted_list.append(left_list[left_list_index])
            left_list_index += 1

    return sorted_list


def merge_sort(nums):  
    # Nếu danh sách là một phần tử duy nhất, trả về nó
    if len(nums) <= 1:
        return nums

    # Sử dụng phân chia sàn để có được điểm giữa, chỉ số phải là số nguyên
    mid = len(nums) // 2

    # Sắp xếp và hợp nhất từng nửa
    left_list = merge_sort(nums[:mid])
    right_list = merge_sort(nums[mid:])

    # Hợp nhất danh sách đã sắp xếp thành danh sách mới
    return merge(left_list, right_list)

list1 = [1, 4, 7]
list2 = [1, 3 ,4]
list3 = [2 ,6]
list = list1 + list2 + list3 
list = merge_sort(list)  
print(list)  
```

#**Thuật toán Quick Sort**

```python

def partition(nums, low, high):  
    pivot = nums[(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while nums[i] < pivot:
            i += 1

        j -= 1
        while nums[j] > pivot:
            j -= 1

        if i >= j:
            return j

        # Nếu một phần tử ở i (bên trái trục) lớn hơn
        # phần tử tại j (bên phải trục), sau đó hoán đổi chúng
        nums[i], nums[j] = nums[j], nums[i]


def quick_sort(nums):  
    # Tạo một hàm trợ giúp sẽ được gọi là đệ quy
    def _quick_sort(items, low, high):
        if low < high:
            # Đây là chỉ mục sau pivot, nơi danh sách của ta được chia
            split_index = partition(items, low, high)
            _quick_sort(items, low, split_index)
            _quick_sort(items, split_index + 1, high)

    _quick_sort(nums, 0, len(nums) - 1)

list1 = [1, 4, 7]
list2 = [1, 3 ,4]
list3 = [2 ,6]
list = list1 + list2 + list3 
quick_sort(list)  
print(list)     
```
